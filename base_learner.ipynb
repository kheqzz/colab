```python
import joblib

# Load the preprocessed data
X_test = joblib.load('X_test_values.pkl')
y_test = joblib.load('y_test_values.pkl')
X_train_resampled = joblib.load('X_train_resampled_values.pkl')
y_train_resampled = joblib.load('y_train_resampled_values.pkl')

print("Data loaded successfully.")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
print(f"X_train_resampled shape: {X_train_resampled.shape}")
print(f"y_train_resampled shape: {y_train_resampled.shape}")


```

```python
# declare model for ensemble learning base learner , Devision Tree,Logistic Regression, Random Forest, Naive Bayes
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

estimators = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('lr', LogisticRegression(random_state=42, max_iter=1000)),
    ('nb', GaussianNB()),
    ('rf', RandomForestClassifier(random_state=42, n_jobs=-1))
    
]
```

```python
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score,accuracy_score,recall_score,precision_score,f1_score

trained_models = {}
#train base learner
for name, model in estimators:
    model.fit(X_train_resampled, y_train_resampled)
    y_pred = model.predict(X_test)
    trained_models[name] = model

    print(f"Model: {name}")
    print(classification_report(y_test, y_pred))

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("Accuracy Score:")
    print(accuracy_score(y_test, y_pred))

    print("Recall Score:")
    print(recall_score(y_test, y_pred))

    print("Precision Score:")
    print(precision_score(y_test, y_pred))

    print("F1 Score:")
    print(f1_score(y_test, y_pred))
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)[:, 1]
        auc_score = roc_auc_score(y_test, y_proba)
        print(f"AUC-ROC Score: {auc_score:.4f}")
    
    #eksport trained model into pkl file
    joblib.dump(model, f'{name}_model.pkl')
    print(f"{name} model saved as {name}_model.pkl\n")
```
