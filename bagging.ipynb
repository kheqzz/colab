```python
import joblib
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import classification_report, f1_score

# 1. Load Data & Base Learners dari pkl
X_train_resampled = joblib.load('X_train_resampled_values.pkl')
y_train_resampled = joblib.load('y_train_resampled_values.pkl')
X_test = joblib.load('X_test_values.pkl')
y_test = joblib.load('y_test_values.pkl')

loaded_dt = joblib.load('dt_model.pkl')
loaded_lr = joblib.load('lr_model.pkl')
loaded_nb = joblib.load('nb_model.pkl')
loaded_rf = joblib.load('rf_model.pkl')

# 2. Definisikan kombinasi Bagging sesuai baris jurnal
bagging_experiments = {
    'Bagging_DT': BaggingClassifier(estimator=loaded_dt, n_estimators=10, random_state=42, n_jobs=-1),
    'Bagging_LR': BaggingClassifier(estimator=loaded_lr, n_estimators=10, random_state=42, n_jobs=-1),
    'Bagging_NB': BaggingClassifier(estimator=loaded_nb, n_estimators=10, random_state=42, n_jobs=-1),
    'Bagging_RF': BaggingClassifier(estimator=loaded_rf, n_estimators=10, random_state=42, n_jobs=-1)
}

# 3. Loop Training & Evaluasi langsung
print("=== EKSPERIMEN BAGGING (COLAB) ===")
for name, model in bagging_experiments.items():
    model.fit(X_train_resampled, y_train_resampled)
    preds = model.predict(X_test)
    print(f"{name} | F1-Score: {f1_score(y_test, preds):.4f}")
    print(classification_report(y_test, preds))
    joblib.dump(model, f'{name}_final.pkl')
```
