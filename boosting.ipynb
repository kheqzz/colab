```python
import joblib
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, f1_score

# (Pastikan baris load pkl data & model dasar sama seperti di atas)
X_train_resampled = joblib.load('X_train_resampled_values.pkl')
y_train_resampled = joblib.load('y_train_resampled_values.pkl')
X_test = joblib.load('X_test_values.pkl')
y_test = joblib.load('y_test_values.pkl')

loaded_dt = joblib.load('dt_model.pkl')
loaded_lr = joblib.load('lr_model.pkl')
loaded_nb = joblib.load('nb_model.pkl')
loaded_rf = joblib.load('rf_model.pkl')
# Definisikan kombinasi Boosting sesuai baris jurnal
boosting_experiments = {
    'Boosting_DT': AdaBoostClassifier(estimator=loaded_dt, n_estimators=50, random_state=42),
    'Boosting_LR': AdaBoostClassifier(estimator=loaded_lr, n_estimators=50, random_state=42, algorithm='SAMME'),
    'Boosting_NB': AdaBoostClassifier(estimator=loaded_nb, n_estimators=50, random_state=42, algorithm='SAMME'),
    'Boosting_RF': AdaBoostClassifier(estimator=loaded_rf, n_estimators=50, random_state=42)
}

print("=== EKSPERIMEN BOOSTING (LOKAL) ===")
for name, model in boosting_experiments.items():
    try:
        model.fit(X_train_resampled, y_train_resampled)
        preds = model.predict(X_test)
        print(f"{name} | F1-Score: {f1_score(y_test, preds):.4f}")
        joblib.dump(model, f'{name}_final.pkl')
    except Exception as e:
        print(f"Gagal training {name} karena: {e}")
```
